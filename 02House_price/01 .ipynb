```python
# missing value处理
missing_data = total1.isnull().sum().sort_values(ascending=False)
missing_percent = ((total1.isnull().sum()) / (total1.isnull().count())).sort_values(ascending=False)
missing_type = total1.dtypes
missing_all = pd.concat([missing_data,missing_percent,missing_type],axis=1,keys=['missing_data','missing_percent','missing_type'])
missing_all.drop(missing_all[missing_data==0].index,inplace=True)
missing_all.sort_values(by='missing_data',ascending=False)
```

```python
total1.drop(missing_all[missing_data>1].index,axis=1,inplace=True)
```

```python
for col in ('GarageCars','GarageArea'):
    total1[col] = total1[col].fillna(0)
for col in ('BsmtFinSF2','BsmtFinSF1','TotalBsmtSF','BsmtUnfSF'):
    total1[col] = total1[col].fillna(0)

total1['KitchenQual'] = total1['KitchenQual'].fillna(total1['KitchenQual'].mode()[0])
total1['Electrical'] = total1['Electrical'].fillna(total1['Electrical'].mode()[0])

total1['Exterior2nd'] = total1['Exterior2nd'].fillna(total1['Exterior2nd'].mode()[0])
total1['Exterior1st'] = total1['Exterior1st'].fillna(total1['Exterior1st'].mode()[0])\

total1['SaleType'] = total1['SaleType'].fillna(total1['SaleType'].mode()[0])

missing_data = total1.isnull().sum().sort_values(ascending=False)
```

```python
total1['MSSubClass'] = total1['MSSubClass'].apply(str)

total1['OverallCond'] = total1['OverallCond'].astype(str)

total1['MoSold'] = total1['MoSold'].astype(str)
total1['YrSold'] = total1['YrSold'].astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder
cols=['BldgType','CentralAir','Condition1','Condition2','Electrical','ExterCond','ExterQual',
     'Exterior1st','Exterior2nd','Foundation','Heating','HeatingQC','HouseStyle','KitchenQual','LandContour',
     'LandSlope','LotConfig','LotShape','Neighborhood','PavedDrive','RoofMatl','PavedDrive','RoofStyle',
      'SaleCondition','SaleType','Street']
le = LabelEncoder()
for i in cols:
    le.fit(list(total1[i].values))
    total1[i] = le.transform(list(total1[i].values))
```

```python
total1['TotalSF'] = total1['TotalBsmtSF'] + total1['1stFlrSF'] + total1['2ndFlrSF']
```

```python
numer_feat = total1.dtypes[total1.dtypes != 'object'].index

skewed_feat = total1[numer_feat].apply(lambda x: (x.dropna()).skew())
skewed_feat = skewed_feat.sort_values(ascending=False)
skewness = pd.DataFrame({'Skew':skewed_feat})
skewness.head()
                     
```

```python
skewness=skewness[abs(skewness)>0.75]
skewness.dropna()

from scipy.special import boxcox1p
skewness_feature = skewness.index
lam=0.15
for i in skewness_feature:
    total1[i] = boxcox1p(total1[i],0.15)
```

```python
preditors=['1stFlrSF', '2ndFlrSF', '3SsnPorch', 'BedroomAbvGr', 'BldgType',
       'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'CentralAir', 'Condition1',
       'Condition2', 'Electrical', 'EnclosedPorch', 'ExterCond', 'ExterQual',
       'Exterior1st', 'Exterior2nd', 'Fireplaces', 'Foundation', 'FullBath',
       'GarageArea', 'GarageCars', 'GrLivArea', 'HalfBath', 'Heating',
       'HeatingQC', 'HouseStyle', 'KitchenAbvGr', 'KitchenQual', 'LandContour',
       'LandSlope', 'LotArea', 'LotConfig', 'LotShape', 'LowQualFinSF',
       'MSSubClass', 'MiscVal', 'MoSold', 'Neighborhood', 'OpenPorchSF',
       'OverallCond', 'OverallQual', 'PavedDrive', 'PoolArea', 'RoofMatl',
       'RoofStyle', 'SaleCondition', 'SaleType', 'ScreenPorch', 'Street',
       'TotRmsAbvGrd', 'TotalBsmtSF', 'WoodDeckSF', 'YearBuilt',
       'YearRemodAdd', 'YrSold', 'TotalSF']

temp1 = total1['source']

total1=pd.get_dummies(total1[preditors])
total1['source']=temp1
total1.shape
```

```python
train = total1[total1['source']=='train']
test = total1[total1['source']=='test']
train.drop(['source'],axis=1,inplace=True)
test.drop(['source'],axis=1,inplace=True)
train.shape,test.shape
```

```python
# 单模型：线性模型、随机森林、GDBR、XGBoost
from sklearn.linear_model import Lasso,BayesianRidge
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.model_selection import cross_val_score,KFold
from sklearn.metrics import mean_squared_error
import xgboost as xgb

lass = Lasso(alpha=0.1)
bayes = BayesianRidge(n_iter=300,tol=0.001,alpha_1=1e-06,alpha_2=1e-06,lambda_1=1e-06)
regr = RandomForestRegressor(max_depth=2,random_state=0)
gbr = GradientBoostingRegressor(loss='ls',learning_rate=0.1,n_estimators=100,subsample=1.0,min_samples_split=2,
                                min_samples_leaf=1,min_weight_fraction_leaf=0.0,max_depth=3,alpha=0.9,)
model_xgb = xgb.XGBRegressor(colsample_bytree=0.4603,gamma=0.0468,learning_rate=0.05,max_depth=3,
                            min_child_weight=1.7817,n_estimators=2200,reg_alpha=0.4640,reg_lambda=0.8571,
                            subsample=0.5213,silent=1,random_state=0,nthread=-1)
```

```python
lass.fit(train,y_train)
bayes.fit(train,y_train)
regr.fit(train,y_train)
gbr.fit(train,y_train)
model_xgb.fit(train,y_train)
```

```python
lass_predict = lass.predict(train)
bayes_predict = bayes.predict(train)
regr_predict = regr.predict(train)
gbr_predict = gbr.predict(train)
model_xgb_predict = model_xgb.predict(train)
```

```python
cv_score_lass = cross_val_score(lass,train,lass_predict,scoring='neg_mean_squared_error',cv=5)
cv_score_bayes = cross_val_score(bayes,train,bayes_predict,scoring='neg_mean_squared_error',cv=5)
cv_score_regr = cross_val_score(regr,train,regr_predict,scoring='neg_mean_squared_error',cv=5)
cv_score_gbr = cross_val_score(gbr,train,gbr_predict,scoring='neg_mean_squared_error',cv=5)
cv_score_model_xgb = cross_val_score(model_xgb,train,model_xgb_predict,scoring='neg_mean_squared_error',cv=5)

lass_score = np.min(cv_score_lass),np.max(cv_score_lass),np.std(cv_score_lass),np.mean(cv_score_lass)
bayes_score = np.min(cv_score_bayes),np.max(cv_score_bayes),np.std(cv_score_bayes),np.mean(cv_score_bayes)
regr_score = np.min(cv_score_regr),np.max(cv_score_regr),np.std(cv_score_regr),np.mean(cv_score_regr)
gbr_score = np.min(cv_score_gbr),np.max(cv_score_gbr),np.std(cv_score_gbr),np.mean(cv_score_gbr)
model_xgb_score = np.min(cv_score_model_xgb),np.max(cv_score_model_xgb),np.std(cv_score_model_xgb),np.mean(cv_score_model_xgb)

total_score = pd.DataFrame({'la_score':lass_score,'bayes_score':bayes_score,
                           'regr_score':regr_score,'gbr_score':gbr_score,
                           'model_xgb_score':model_xgb_score},
                          index=['min','max','std','mean'])
```
